```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/home/thenielfarias/Market_Basket_Optimisation.csv', header = None)
```

```python
data
```

```python
transactions = []

for i in range(0, 7501):
  transactions.append([str(data.values[i,j]) for j in range(0, 20)])
```

```python
# Training the model on the dataset

from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3,
                min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
results
```

```python
# Apriori results

def apriori_inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

df_results_apriori = pd.DataFrame(apriori_inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support',
                                                                       'Confidence', 'Lift'])
```

```python
df_results_apriori
```

```python
df_results_apriori.nlargest(n = 10, columns = 'Lift')
```

```python
# Eclat results

def eclat_inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))

df_results_eclat = pd.DataFrame(eclat_inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

```python
df_results_eclat.nlargest(n = 10, columns = 'Support')
```
